```python
data1=data.drop(['Race','Marital Status','differentiate','A Stage'],axis=1)
data1
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.ensemble import GradientBoostingClassifier
```

```python
data1.columns
```

```python
from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()
data1['Status'] = data1['Status'].map({'Alive':1,'Dead':0})
data1['T Stage '] = data1['T Stage '].map({'T1':0,'T2':1,'T3':2})
data1['N Stage'] = data1['N Stage'].map({'N1':0,'N2':1,'N3':2})
data1['Estrogen Status'] = data1['Estrogen Status'].map({'Positive':1,'Negative':0})
data1['Progesterone Status'] = data1['Progesterone Status'].map({'Positive':1,'Negative':0})
```

```python
data1
```

```python
data1.Grade.value_counts()
```

```python
data1['Grade']=encoder.fit_transform(data1['Grade'])
```

```python
data1['6th Stage'] = data1['N Stage'].map({'IIA':0,'IIIA':1,'IIIC':2})
```

```python
data1
```

```python
x=data1.drop('Status',axis=1)
y=data1['Status']
```

```python
x.isnull().sum()
```

```python
x['6th Stage']=x['6th Stage'].fillna(0)
x['T Stage ']=x['T Stage '].fillna(0)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_train_scaled=scaler.fit_transform(x_train)
x_test_scaled=scaler.transform(x_test)
```

```python
model=GradientBoostingClassifier()

model.fit(x_train_scaled,y_train)
```

```python
model_pred=model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report
accuracy=accuracy_score(y_test,model_pred)
print(f'accuracy is: {accuracy}')
```

```python
classification=classification_report(y_test,model_pred)
print(classification)
```
